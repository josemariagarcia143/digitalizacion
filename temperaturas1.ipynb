# Ejercicio para convertir temperaturas Celsius a Fahrenheit

```python
celsius = float(input('Dime la temperatura en ºC: '))
fahrenheit = celsius * (9/5) + 32
kelvin = celsius + 273.15
print('La temperatura de', celsius,'ºC =',fahrenheit,'ºF =',kelvin,'K')
```
